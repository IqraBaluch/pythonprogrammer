````python
import pandas as pd

df = pd.read_csv("ransom.csv")



# Display the first five rows of the DataFrame and determine which columns to plot.

df.head()

````

````python
from matplotlib import pyplot as plt
#Create a scatter plot of the data


plt.scatter(df.letter_index, df.requency)

plt.show()
````

# Modifying a scatterplot

Change the color of the points to 'red'.

````python
from matplotlib import pyplot as plt
#Create a scatter plot of the data


plt.scatter(df.letter_index, df.requency, color = "red")

plt.show()
````

````python
# Change the marker shape to square.
plt.scatter(df.letter_index, df.requency, color = "red", marker="s")

plt.show()
````

````python
# Change the transparency of the scatterplot to 0.1.

plt.scatter(df.letter_index, df.requency, color = "red", marker="s", alpha = 0.1)

plt.show()
````

## Adding error Bars



## Stacked Bar chart for comparision


````python
# Build a simple bar chart

plt.bar(df.letter_index, df.requency)

plt.show()
````

````python
#  add error bars to the bar chart.
plt.bar(df.letter_index, df.requency, yerr = df.requency)

plt.show()
````

````python
# Label that bar plot

plt.bar(df.letter_index, df.requency, yerr = df.requency, label = "Ransom Data")

plt.legend()

plt.show()
````

````python
# Separate the data | use bottom

plt.bar(df.letter_index, df.requency, yerr = df.requency, label = "Ransom Data", bottom = df.letter_index)

plt.legend()

plt.show()
````

# Making a histogram

A histogram visualizes the distribution of values in a dataset. 

To create a histogram we place each peice if data into  bin. 




ploting a histogram in matplotlib is easy, use the command ```plt.hist()``` this function takes only one positional argument. 


### Changing Bins

```plt.hist(data, bins = 40)```

#### Range
use keyword range to set minumum and mximum value for histogram



### Normalization




``` density = True ```



## Modifying histograms



````python
# Create a histogram
plt.hist(df['requency'])

plt.show()
````

````python
# Change the number of bins to 50.
plt.hist(df['requency'], bins = 50)

plt.show()

````

````python
# Change the range to start at 5 and end at 35.
plt.hist(df['requency'], bins = 50, range = (5,35))

plt.show()


````

````python
# Normalize your histogram so that the sum of the bins adds to 1.

plt.hist(df['requency'], bins = 50, range = (1,13), density = True)

plt.show()
````

````python
# Label the x-axis, the y-axis (Frequency), and add the title.

plt.hist(df['requency'], bins = 50, range = (1,14), density = True)

# Label plot


plt.xlabel("Index")
plt.ylabel("Frequency")
plt.title("Ransom Data")

plt.show()

````
