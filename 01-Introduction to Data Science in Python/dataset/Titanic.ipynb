# Scatter plot

```python
# make a scatter plot
plt.scatter(titanic.Age, titanic.Fare)

#give a tittle to plot
plt.title("Age VS Fare")

#x-label
plt.xlabel("Ages")
plt.ylabel("Fare")


plt.show()
```

```python
plt.scatter(titanic.Sex, titanic.Age)

#give a tittle to plot
plt.title("Sex VS Age")

#x-label
plt.xlabel("Sex")
plt.ylabel("Age")


plt.show()
```

# Bar Chart

```python
plt.bar( titanic.Sex, titanic.Fare, label = "Genders")


#give a tittle to plot
plt.title("Gender V/S survived")

#x-label
plt.xlabel("Sex")
plt.ylabel("Number of passengers")

plt.legend()


plt.show()
```

```python
plt.bar( titanic.Embarked, titanic.Fare, label = "embark history")
plt.bar( titanic.Sex, titanic.Age, label = "age history", bottom = titanic.Age)

#give a tittle to plot
plt.title("Embark V/S Fare")

#x-label
plt.xlabel("Embark")
plt.ylabel("Fare")

plt.legend()

plt.show()
```

# Histogram


```python
plt.hist( titanic.Age, bins = 100, range = (5, 79), density = True)

#give a tittle to plot
plt.title("Ages of Customers")

#x-label
plt.xlabel("Age")


plt.show()
```
