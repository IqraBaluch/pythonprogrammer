````python
# From matplotlib, import pyplot under the alias plt
import pandas as pd
from matplotlib import pyplot as plt
````

# Adding text to plots




looking at the plot we are unable to understand it. 

when we share our plot we must use labels to make it clear what we have plotted and why is it important.  

1. Firts add axes labels

we can label the x-axis using the function ```plt.xlabel()```



2. similarly we can label the y label too.


``` plt.ylabel()```



3. Finally we can give our plot a title. 


``` plt.title(_)```


### Legend 

if we have multiple lines in a plot, we will add a legend.

There are two steps to add a legend

1. we must add the keyword argument label


2. add legend function

``` plt.legend() ```

### Arbitarary Text

sometimes we only want to add quick note directly into our plot. 
we can do this by using the ```plt.text()``` this function takes 3 arguments. the x-coordinate where we want to put the text. the y-coordinate where we want to put the text, and the string.





````python
plt.plot(ransom.letter_index, ransom.requency)


plt.text(7,9, "Usually low H frequency!")

plt.show()
````

## Adding a legend


Using the keyword label, label ransom plot as "Ransom".

````python
rnsm = pd.read_csv("ransom.csv")

rnsm.head()

````

````python

plt.plot(rnsm.letter_index, rnsm.requency, label = "Frequency")



#Add a descriptive title to the chart.

plt.title("Ransom clues Chart")

# Add a label for the y-axis

plt.ylabel("Frequency")


#Nothing is displaying yet! Add a command to make the legend display.
plt.legend()

plt.show()
````

# Adding floating text

 Place the annotation "high frequency" at the point (2.5, 80).

````python
plt.plot(rnsm.letter_index, rnsm.requency, label = "Frequency")



#Add a descriptive title to the chart.

plt.title("Ransom clues Chart")

# Add a label for the y-axis

plt.ylabel("Frequency")

#annotation
plt.text(2.5, 11, "High frequency")
plt.text( 15, 8, "low frequency")

#Nothing is displaying yet! Add a command to make the legend display.
plt.legend()

plt.show()
````

# Playing with styles

'fivethirtyeight' - Based on the color scheme of the popular website

'grayscale' - Great for when you don't have a color printer!

'seaborn' - Based on another Python visualization library
'classic' - The default color scheme for Matplotlib

````python
plt.plot(rnsm.letter_index, rnsm.requency, label = "Frequency")



#Add a descriptive title to the chart.

plt.title("Ransom clues Chart")

# Add a label for the y-axis

plt.ylabel("Frequency")


plt.style.use("fivethirtyeight")

plt.legend()

plt.show()
````

````python

plt.plot(rnsm.letter_index, rnsm.requency, label = "Frequency")

#Add a descriptive title to the chart.

plt.title("Ransom clues Chart")

# Add a label for the y-axis

plt.ylabel("Frequency")

plt.style.use("fivethirtyeight")

plt.legend()


plt.show()
````

````python
plt.plot(rnsm.letter_index, rnsm.requency, label = "Frequency")


#Add a descriptive title to the chart.

plt.title("Ransom clues Chart")

# Add a label for the y-axis

plt.ylabel("Frequency")

plt.style.use("grayscale")

plt.legend()


plt.show()
````

````python
# # Change the style to ggplot/

plt.plot(rnsm.letter_index, rnsm.requency, label = "Frequency")


#Add a descriptive title to the chart.

plt.title("Ransom clues Chart")

# Add a label for the y-axis

plt.ylabel("Frequency")

plt.style.use("ggplot")

plt.legend()


plt.show()
````
